```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd

graph_data = pd.read_excel('time_insertion_deletion_AVL.xlsx')
df_graph = pd.DataFrame(graph_data)

df_insertion = df_graph['time_insertion']
df_deletion = df_graph['time_deletion']

x_axis = list(np.arange(1,23,1))

df_deletion.plot(x = x_axis, y = 'time_deletion', kind = 'line')
plt.xlabel('No of elements')
plt.ylabel('Time in seconds')
plt.title('Time for deletion of key from AVL Tree')
plt.show()
```
